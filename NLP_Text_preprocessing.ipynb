```python
#cleaning text
```

```python
stop_words=stopwords.words('english')
```

```python
lemmatizer=WordNetLemmatizer()
lemmatizer
```

```python
def cleaningtext(text):
  text = text.lower()
  text = contractions.fix(text)
  text = re.sub(r'[^a-zA-Z\s]','',text)
  text = " ".join ([word for word in text.split() if word.isalpha()])

  token = word_tokenize(text)

  output_text = " ".join([lemmatizer.lemmatize(word) for word in token if word not in stop_words])
  return output_text
```

```python
x['Review_cleaned'] = x['Review'].apply(cleaningtext)
```

```python
from concurrent.futures import ThreadPoolExecutor
```

```python
def single_translate(text):
    if not isinstance(text,str) or text.strip()=="": return text

    if all(ord(char)<128 for char in text):
        return text

    try:
        return GoogleTranslator(source='auto',target='en').translate(text)
    except:
        return text


def parallel_translate(sentence,max_workers=10):
    with ThreadPoolExecutor(max_workers=max_workers) as pool:
        return list(pool.map(single_translate,sentence))
```

```python
x['Review_cleaned_text_translated'] = parallel_translate(x['Review_cleaned_text'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y_Sentiment,test_size=0.2,random_state=42)
```

```python
vectorize = TfidfVectorizer()
x_train_vector = vectorize.fit_transform(x_train)
x_test_vector = vectorize.transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train_vector,y_train)
```

```python
#model evaluvation

y_train_predict = model.predict(x_train_vector)
y_test_predict = model.predict(x_test_vector)

print("Training Accuracy:",accuracy_score(y_train,y_train_predict))
print("Testing Accuracy:",accuracy_score(y_test,y_test_predict))

# classfication matrix
print("Classification Report - Training Data:")
print(classification_report(y_train, y_train_predict))

print("\nClassification Report - Testing Data:")
print(classification_report(y_test, y_test_predict))
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, hamming_loss
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import make_scorer
from sklearn.ensemble import RandomForestClassifier
```

```python
df['Topic_list'] = df['Topic'].apply(lambda x: [t.strip() for t in str(x).split(',')])
```

```python
x = df['Review_cleaned_text_translated']
y_topic = df['Topic_list']
```

```python
mlb = MultiLabelBinarizer()
y_encoded = mlb.fit_transform(y_topic)
```

```python
y_encoded
```

```python
mlb.classes_
```

```python
pd.DataFrame(y_encoded,columns=mlb.classes_)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y_encoded,test_size=0.2,random_state=42)
```

```python
vectorize = TfidfVectorizer()
x_train_vector = vectorize.fit_transform(x_train)
x_test_vector = vectorize.transform(x_test)
```

```python
model = OneVsRestClassifier(LogisticRegression())
model.fit(x_train_vector,y_train)
```

```python
# model evaluation

y_train_predict = model.predict(x_train_vector)
y_test_predict = model.predict(x_test_vector)

print("Training Accuracy:",accuracy_score(y_train,y_train_predict))
print("Testing Accuracy:",accuracy_score(y_test,y_test_predict))

# classification

print("Classification Report - Training Data:")
print(classification_report(y_train, y_train_predict, target_names=mlb.classes_))

print("Classification Report - Testing Data:")
print(classification_report(y_test, y_test_predict, target_names=mlb.classes_))

#k_fold

k_fold = KFold(n_splits=10, shuffle=True, random_state=42)
cv = cross_val_score(model, x_train_vector, y_train, cv=k_fold, scoring='accuracy')
print("Cross Validation Score:", cv.mean())

# hamming loss

hamming_loss_train = hamming_loss(y_train, y_train_predict)
hamming_loss_test = hamming_loss(y_test,  y_test_predict)

print("Hamming Loss - Training Data:", hamming_loss_train*100)
print("Hamming Loss - Testing Data:", hamming_loss_test*100)

# hamming score

hamming_score = make_scorer(hamming_loss,greater_is_better=False)
cv = cross_val_score(model, x_train_vector, y_train, cv=k_fold, scoring=hamming_score)
print("Hamming Score:", -cv.mean()*100)
```
